```python
def joiner(lst, id_):

    res_df = lst[0].set_index(id_)
    
    for df in lst[1:]:
        res_df = res_df.join(df.set_index(id_), how='left', on = id_)
    
    
    return res_df
```

```python
def value_counter(lst):

    names  = [df.name for df in lst]
    lst = [df.iloc[:,1:] for df in lst]
    lst_len = len(lst)
    
    for i in range(lst_len):
        print('\n')
        print(f'–ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ "{names[i]}":')
        print('\n')
        for i in df.select_dtypes(exclude=np.number).columns:
            print(f'–ö–æ–ª–æ–Ω–∫–∞ {i}:')
            print(df[i].unique().tolist())
            print('-'*100)
        print('\n')
```

```python
def c_plot(df, tuple_, figsize=(12,6)):

    tuple_len = len(tuple_)

    cols = [i[0] for i in tuple_]
    
    fig, ax = plt.subplots(tuple_len, 1, figsize=figsize)

    for i in range(tuple_len):
        sns.countplot(df, y=cols[i], palette='husl', alpha=0.7,  ax = ax[i], legend=False);

        #ax[i].tick_params(labelrotation=30)
        ax[i].spines['top'].set_visible(False)
        ax[i].spines['right'].set_visible(False)
        #ax[i].spines['left'].set_visible(False)

        ax[i].set_title(tuple_[i][1],loc='left')
        ax[i].set_xlabel(tuple_[i][2])
        ax[i].set_ylabel(tuple_[i][3])

    fig.align_ylabels()        
    fig.subplots_adjust(hspace=1.5)
```
